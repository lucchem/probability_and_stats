```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
n_games = 100000
outcomes = np.array([])

for game in range(n_games):
    outcomes = np.append(outcomes, np.random.random_integers(1,6,100).sum() - np.random.random_integers(0,1,600).sum())
```

```python
histdata = (outcomes - (100*3.5 - 600*0.5))/np.sqrt(100*35/12+600*0.25)

plt.hist(histdata, bins = 100, histtype = 'step', density = True)
h = np.linspace(-4,4,100)
plt.plot(h,sts.norm.pdf(h))
plt.show()
```
